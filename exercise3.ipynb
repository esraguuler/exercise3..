```python
# Kütüphanelerin eklenmesi
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Veri kümesinin yüklenmesi
df = pd.read_csv('C:/Users/esrag/OneDrive/Masaüstü/train_and_test2.csv')
```

```python
# Veri kümesiyle ilgili temel bilgilerin görüntülenmesi
print(df.info())
```

```python
# Eksik değerlerin işlenmesi
df = df.dropna() 
```

```python
# Temel istatistiklerin keşfedilmesi
print(df.describe())
```

```python
# Değişkenlerin dağılımının görselleştirilmesi
for column in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Korelasyon matrisi ve ısı haritası kullanarak korelasyonların belirlenmesi
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
